````python
class Perceptron:
    def __init__(self,learning_rate=0.1,epochs=10):
        self.lr=learning_rate
        self.epochs=epochs
        self.weights=None
        self.bias=None

    def step_function(self,z):
        return 1 if z>=0 else 0

    def fit(self,X,y):
        # number of features
        n_features=len(X[0])

        # initialize weights and bias
        self.weights=[0.0]*n_features
        self.bias=0.0

        # training loop
        for epoch in range(self.epochs):
            print(f"Epoch {epoch + 1}")
            for i in range(len(X)):
                # weighted sum
                z=sum(w*x for w,x in zip(self.weights,X[i]))+self.bias

                # prediction
                y_hat=self.step_function(z)

                # error
                error=y[i]-y_hat

                # update weights and bias
                for j in range(n_features):
                    self.weights[j] +=self.lr*error*X[i][j]
                self.bias +=self.lr*error

                print(
                    f"  Input: {X[i]}, "
                    f"Prediction: {y_hat}, "
                    f"Error: {error}, "
                    f"Weights: {self.weights}, "
                    f"Bias: {self.bias}"
                )

    def predict(self,X):
        predictions=[]
        for x in X:
            z=sum(w*xi for w,xi in zip(self.weights,x))+self.bias
            predictions.append(self.step_function(z))
        return predictions
````

````python
# AND gate dataset
X=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

y=[0,0,0,1]
````

````python
p=Perceptron(learning_rate=0.1,epochs=10)
p.fit(X,y)
````

````python
predictions=p.predict(X)
print("Final predictions:",predictions)
````


# 1. Gradient Descent (Optimization Methods)

### Definition

**Gradient Descent** is a method to **minimize a loss function** by moving parameters (weights) in the direction of **steepest decrease**.

[
w \leftarrow w - \eta \frac{\partial L}{\partial w}
]

---

## Common Types of Gradient Descent

### 1. Batch Gradient Descent

**Uses:** all training data
**Update:** once per epoch

✔ Stable convergence
❌ Slow for large datasets

**Used when:** small datasets, theoretical clarity

---

### 2. Stochastic Gradient Descent (SGD)

**Uses:** one data point at a time
**Update:** very frequent

✔ Fast, escapes local minima
❌ Noisy updates

**Used when:** very large datasets, online learning

---

### 3. Mini-Batch Gradient Descent (MOST COMMON)

**Uses:** small batches (e.g. 32, 64)

✔ Balance of speed + stability
✔ GPU-friendly

**Used when:** almost all deep learning models

---

### 4. Momentum

Adds memory of past gradients

✔ Faster convergence
✔ Reduces oscillation

**Used when:** deep networks, ravines in loss surface

---

### 5. RMSProp

Adapts learning rate per parameter

✔ Good for non-stationary problems

---

### 6. Adam (MOST POPULAR)

Momentum + RMSProp combined

✔ Fast
✔ Robust
✔ Minimal tuning

**Used when:** default optimizer in practice

---

# 2. Loss Functions (What We Minimize)

### Definition

A **loss function** measures **how wrong a prediction is**.

[
L(y, \hat{y}) = \text{badness of prediction}
]

---

## Common Loss Functions

### 1. Mean Squared Error (MSE)

[
L = (y - \hat{y})^2
]

**Used for:** regression
✔ Simple
❌ Sensitive to outliers

---

### 2. Mean Absolute Error (MAE)

[
L = |y - \hat{y}|
]

**Used for:** regression
✔ Robust to outliers
❌ Gradient not smooth at zero

---

### 3. Binary Cross-Entropy (Log Loss)

[
L = -[y\log(\hat{y}) + (1-y)\log(1-\hat{y})]
]

**Used for:** binary classification
✔ Works with sigmoid
✔ Probabilistic interpretation

---

### 4. Categorical Cross-Entropy

**Used for:** multi-class classification
✔ Works with softmax

---

### 5. Sparse Categorical Cross-Entropy

Same as above, but labels are integers

---

### 6. Hinge Loss

**Used for:** Support Vector Machines
✔ Maximizes margin

---

### 7. Perceptron Loss (Historical)

[
L = \max(0, -y(w \cdot x))
]

**Used for:** classic perceptron
❌ Not smooth

---

# 3. Activation Functions (How Neurons Decide)

### Definition

An **activation function** transforms the weighted sum into the neuron’s **output**.

[
\hat{y} = f(w \cdot x + b)
]

---

## Common Activation Functions

### 1. Step Function

**Output:** 0 or 1

✔ Simple
❌ Not differentiable

**Used for:** perceptron (historical)

---

### 2. Sigmoid

[
\sigma(z) = \frac{1}{1 + e^{-z}}
]

**Output:** (0, 1)

✔ Probabilities
❌ Vanishing gradients

**Used for:** binary classification (output layer)

---

### 3. Tanh

**Output:** (−1, 1)

✔ Zero-centered
❌ Still vanishing gradients

**Used for:** older neural networks

---

### 4. ReLU (Rectified Linear Unit)

[
f(z) = \max(0, z)
]

✔ Fast
✔ Sparse activation
❌ Dead neurons

**Used for:** hidden layers (MOST COMMON)

---

### 5. Leaky ReLU

Fixes dead ReLU problem

**Used for:** deeper networks

---

### 6. ELU / SELU

Smoothed ReLU variants

**Used for:** self-normalizing networks

---

### 7. Softmax

[
\hat{y}_i = \frac{e^{z_i}}{\sum e^{z_j}}
]

**Output:** probability distribution

✔ Multi-class classification

**Used for:** final layer in multi-class problems

---

# 4. How They Work Together (Big Picture 🔥)

Input
 → Linear (w·x + b)
