```python
import numpy as np
import pandas as pd

# number of samples
n = 200

# generate x values
X = np.random.uniform(-4, 4, n)

# generate U-shaped curve
Y = X**2 + np.random.normal(0, 1.5, n)

# create labels (two classes)
label = (Y > 6).astype(int)

# create dataframe
df = pd.DataFrame({
    "X": X,
    "Y": Y,
    "label": label
})

# save to csv
df.to_csv("ushape.csv", index=False)

print("ushape.csv created successfully")
```
