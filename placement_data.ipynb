```python
import pandas as pd
import numpy as np
np.random.seed(42)
n=200
```

```python
cgpa=np.random.normal(7.8,0.8,n).clip(4.0,10.0)
package=[]
for gpa in cgpa:
    base=2+(gpa-4)*1.8
    noice=np.random.normal(0,0.8,1)[0]
    pkg=max(1.0,base+noice)
    package.append(round(pkg,1))
df=pd.DataFrame({
    'cgpa':[round(g,2) for g in cgpa],
    'package':package
})
df.to_csv('placement.csv', index=False)
print("Dataset created: placement_data.csv")
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
print(f"CGPA range: {df.cgpa.min():.2f} - {df.cgpa.max():.2f}")
print(f"Package range: {df.package.min()} - {df.package.max()} LPA")
```
