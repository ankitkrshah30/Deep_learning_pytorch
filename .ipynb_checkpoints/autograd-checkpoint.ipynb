# Basic of Diffrentiation


```python
import math
def dy_dx(x):
    return 2*x
def dz_dy(y):
    return math.cos(y)
def du_dz(z):
    return math.exp(z)
def du_dx(x):
    y=x**2
    z=math.sin(y)
    return dy_dx(x)*dz_dy(y)*du_dz(z)
#y=x^2
#z=sin y
#u=e^z
du_dx(2)
```

# Using Autograd

```python
 import torch
```

```python
x=torch.tensor(2.0,requires_grad=True)
#only floating point numbers are required
x
```

```python
y=x**2
y
```

```python
z=torch.sin(y)
z
```

```python
u=torch.exp(z)
u
```

```python
u.backward()
```

```python
u.grad
```

```python
x.grad
```
