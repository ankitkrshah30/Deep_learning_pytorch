```python
#add
a+b
```

```python
#sub
a-b
```

```python
#multiply
a*b
```

```python
#division
a/b
```

```python
#power
a**b
```

```python
#mod
a%b
```

```python
c=torch.tensor([1,-2,3,-4])
```

```python
#abs
torch.abs(c)
```

```python
#negative
torch.neg(a)
```

```python
#round
torch.round(a)
```

```python
#ceil
torch.ceil(a)
```

```python
#floor
torch.floor(a)
```

```python
#clamp
torch.clamp(a,min=2,max=3)
```

## 3. Reduction operation

```python
e=torch.randint(size=(2,3),low=0,high=10,dtype=torch.float64)
e
```

```python
#sum
torch.sum(e)
```

```python
#sum along columns
torch.sum(e,dim=0)
```

```python
#sum alomg rows
torch.sum(e,dim=1)
```

```python
#mean
torch.mean(e)
```

```python
torch.mean(e,dim=0)
```

```python
#median
torch.median(e)
```

```python
# max and min
print(torch.max(e))
print(torch.min(e))
```

```python
# product
torch.prod(e)
```

```python
# standard deviation
torch.std(e)
```

```python
#variance
torch.var(e)
```

```python
#argmax
torch.argmax(e)
```

```python
#argmin
torch.argmin(e)
```

## 4. Matrix operations

```python
f=torch.randint(size=(2,3),low=0,high=10)
g=torch.randint(size=(3,2),low=0,high=10)
print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
vector1=torch.tensor([1,2])
vector2=torch.tensor([3,4])
#dot product
torch.dot(vector1,vector2)
```

```python
#transpose
torch.transpose(f,0,1)
```

```python
#determinant
x=torch.randint(size=(3,3),low=0,high=10,dtype=torch.float32)
print(x)
torch.det(x)
```

```python
#inverse
torch.inverse(x)
```

## 5. Comparision operations

```python
i=torch.randint(size=(2,3),low=0,high=10)
j=torch.randint(size=(2,3),low=0,high=10)
print(i)
print(j)
```

```python
# greater than
i>j
```

```python
# less than
i<j
```

```python
# equal to
i == j
```

```python
# not equal to
i != j
```

```python
# greater than equal to

# less than equal to
```
